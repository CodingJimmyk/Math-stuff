Two wave function(square&triangle) and Fourier series with 2, 5, 8, and 100 terms

```python
import numpy as np

import matplotlib.pyplot as plt

from scipy.signal import square

from scipy.integrate import quad

from math import* 
```

```python
def plot_fourier(n, x, y, wave_name, wave_function):
  if wave_function == 0:
    fc=lambda x:square(x)*cos(i*x)  
    fs=lambda x:square(x)*sin(i*x)
  else:
    fc=lambda x:sawtooth(x+1/2*np.pi, 0.5)*cos(i*x)  #i :dummy index
    fs=lambda x:sawtooth(x+1/2*np.pi, 0.5)*sin(i*x)

  
  An=[] 

  Bn=[]

  for i in range(n):


    An.append(0)

  for i in range(n):

    bn=quad(fs,-np.pi,np.pi)[0]*(1.0/np.pi)

    Bn.append(bn) 

  for i in range(n):


    if i==0.0:
      
      sum=0

    else:

      sum=sum+(An[i]*np.cos(i*x)+Bn[i]*np.sin(i*x))

  plt.plot(x,sum,'g')

  plt.plot(x,y,'r--')

  plt.title(f"fourier series for {wave_name} wave with {n} terms")

  plt.show()
```

```python
x = np.arange(-2*np.pi, 4*np.pi, 0.001)
y1 = square(x)
y2 = sawtooth(x+1/2*np.pi, 0.5)  #triangle wave function 
# 0 for square, 1 for traingle
plot_fourier(2, x, y1, 'square', 0)
plot_fourier(5, x, y1, 'square', 0)
plot_fourier(8, x, y1, 'square', 0)
plot_fourier(100, x, y1, 'square', 0)
plot_fourier(2, x, y2, 'triangle', 1)
plot_fourier(5, x, y2, 'triangle', 1)
plot_fourier(8, x, y2, 'triangle', 1)
plot_fourier(100, x, y2, 'triangle', 1)
```
